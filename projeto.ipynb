Como acessar uma página com o selenium

```python
from selenium import webdriver

navegador = webdriver.Chrome()
```

```python
# Abrindo o site da Hashtag Treinamentos
navegador.get("https://www.hashtagtreinamentos.com.br/")
navegador.maximize_window()
```

Acessando uma página localmente

```python
import os

caminho = os.getcwd()
arquivo = r"Pagina Hashtag.html"

navegador.get(caminho + "/" + arquivo)
navegador.maximize_window()
```

Selecionar elemento na página

```python
from selenium.webdriver.common.by import By

navegador.get("https://www.hashtagtreinamentos.com.br/")
navegador.switch_to.window(navegador.window_handles[0])
navegador.maximize_window()
```

Clicando no botão de login

```python
botao_login = navegador.find_element(By.XPATH, "/html/body/header/div/div/a[1]")
botao_login.click()
```

Escrevendo no campo de e-mail

```python
campo_email = navegador.find_element(By.ID, "ap-email")
campo_email.send_keys("calazanssergioherique@gmail.com")
```

Clicando no botão de acessar o portal

```python
botao_portal = navegador.find_element(By.ID, "ap-submit")
botao_portal.click()
```
