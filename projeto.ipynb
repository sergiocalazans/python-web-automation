Como acessar uma página com o selenium

```python
from selenium import webdriver

navegador = webdriver.Chrome()
```

```python
# Abrindo o site da Hashtag Treinamentos
navegador.get("https://www.hashtagtreinamentos.com.br/")
navegador.maximize_window()
```

Acessando uma página localmente

```python
import os

caminho = os.getcwd()
arquivo = r"Pagina Hashtag.html"

navegador.get(caminho + "/" + arquivo)
navegador.maximize_window()
```
